### 2.4 LSTM 모델링

#### 2.4.1 개별 함수 선언

```python
def set_data_frame(df_stock) :
    stock_4d = len(df_stock['group'].unique())
    stock_3d = int(len(df_stock['ticker'].unique())/ stock_4d)
    stock_2d = int(len(df_stock) / stock_4d / stock_3d)
    
    stock_array = np.array(df_stock.drop(['group', 'ticker'], axis=1, inplace=False))
    stock_array = stock_array.reshape(stock_4d, stock_3d, stock_2d, -1)
    stock_array_set = stock_array.copy()
    
    return stock_array_set


def min_max_scaler(df_def):
    scaler_feature = MinMaxScaler()
    scaler_label = MinMaxScaler()          
    
    scaled_feature = scaler_feature.fit_transform(df_def[:,:-1])
    scaled_label = scaler_label.fit_transform(df_def[:, -1].reshape(-1,1))
    scaled_stock = np.hstack([scaled_feature, scaled_label])
    
    return scaled_stock, scaler_feature, scaler_label


def divide_set(scaled_stock, train_size=60, test_size=20):
    train_data_set = scaled_stock[:-test_size]
    test_data_set = scaled_stock[-test_size -train_size:]
    
    return train_data_set, test_data_set


def split_sequences(sequences, n_steps_in=60, n_steps_out=20, step_size=1):
    X, y = list(), list()
    
    for i in range(len(sequences)):
        i_step = i * step_size
        end_ix = i_step + n_steps_in
        out_end_ix = end_ix + n_steps_out
        
        if out_end_ix > len(sequences):
            break
        
        seq_x, seq_y = sequences[i_step:end_ix, :], sequences[end_ix:out_end_ix, -1]
        X.append(seq_x)
        y.append(seq_y)
    
    return np.array(X), np.array(y)


def splitting_train_valid (train, test, test_size=0.2):
    X_train, X_val, y_train, y_val = train_test_split(train, test, test_size=test_size , random_state=100)
    
    return X_train, X_val, y_train, y_val


def set_lstm (df_def, n_steps_out):
    model = Sequential()
    model.add(LSTM(100, 
                   input_shape=(df_def.shape[1], df_def.shape[2]), 
                   activation='relu', 
                   return_sequences=True)
              )
    model.add(Dropout(0.1))
    model.add(LSTM(50, activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(n_steps_out))
    model.compile(optimizer = 'adam', loss = 'Huber', metrics=['mse'])  
    
    return model

    
def model_fit (model, X_train, X_val, y_train, y_val, X_test, y_test, scaler_label, i, j):  
    model = model
    early_stop = EarlyStopping(monitor='val_loss', patience=10)
    
    if i == 0:
        group = "A_group"
    elif i == 1:
        group = "B_group"
    elif i == 2:    
        group = "C_group"
    elif i == 3:
        group = "D_group"    
    
    model_h5_dir = './save_model/LSTM_h5'
    cur_dir = os.getcwd() #현재 작업디렉토리
    model_h5_path = os.path.join(cur_dir, model_h5_dir)
    model_h5_file = os.path.join(model_h5_path, f'LSTM-{group}-{j:02d}.h5')
    os.makedirs(model_h5_path, exist_ok=True)   
    history = model.fit(X_train, y_train, 
                    epochs=200, 
                    steps_per_epoch=10,
                    validation_data=(X_val, y_val), 
                    callbacks=[early_stop])
    model.save(model_h5_file, save_format='h5')
    y_pred = model.predict(X_test)
    y_pred_origin = scaler_label.inverse_transform(y_pred.reshape(-1,1))
    y_test_origin = scaler_label.inverse_transform(pd.DataFrame(y_test.reshape(-1,1)))
    
    return history, y_pred, y_pred_origin, y_test_origin


def train_process(hist, col, low, i, j):
    rows = low
    columns = col
    image_index = (j*4)+(i+1)
    
    if i == 0:
        group = "A_group"
    elif i == 1:
        group = "B_group"
    elif i == 2:    
        group = "C_group"
    elif i == 3:
        group = "D_group"      
    
    title = f'img_train_result_{group}_{j:02d}'
    plt.figure()
    plt.plot(hist.history['loss'], color = 'red', label='train loss')
    plt.plot(hist.history['val_loss'], color = 'blue', label='val loss')
    plt.title(title)
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.grid()
    plt.legend(loc = 'upper right')
    plt.savefig('./image/'+ str(title))
    plt.tight_layout() 
    plt.show()
    
    return 


def Predictive_evaluation_scaled( y_pred, y_test, col, low, i, j):
    rows = low
    columns = col
    image_index = (j*4)+(i+1)
    
    if i == 0:
        group = "A_group"
    elif i == 1:
        group = "B_group"
    elif i == 2:    
        group = "C_group"
    elif i == 3:
        group = "D_group"      
    
    title = f'img_evaluation_scaled_{group}_{j:02d}'
    plt.figure()
    plt.plot(y_test.reshape(-1,1), color = 'red', label='real stroc scaled price')
    plt.plot(y_pred.reshape(-1,1), color = 'blue', label='predicted stroc scaled price')
    plt.title(title)
    plt.xlabel('time')
    plt.ylabel('stock scaled price')
    plt.grid()
    plt.xticks(range(20), range(1,21))
    plt.yticks(np.arange(0, 1, 0.1))
    plt.legend(loc = 'lower right')
    plt.savefig('./image/'+ str(title))
    plt.tight_layout() 
    plt.show()
    
    return


def Predictive_evaluation(y_pred_origin, y_test_origin, col, low, i, j):
    rows = low
    columns = col
    image_index = (j*4)+(i+1)
    
    if i == 0:
        group = "A_group"
    elif i == 1:
        group = "B_group"
    elif i == 2:    
        group = "C_group"
    elif i == 3:
        group = "D_group"      
    
    title = f'img_evaluation_{group}_{j:02d}'
    plt.figure()
    plt.plot(y_test_origin, color = 'red', label='real stroc price')
    plt.plot(y_pred_origin, color = 'blue', label='predicted stroc price')
    plt.title(title)
    plt.xlabel('time')
    plt.ylabel('stock price')
    plt.xticks(range(20), range(1,21))
    max_price = max([max(y_pred_origin)[0], max(y_test_origin)[0]])
    graid = np.math.ceil((max_price/10)/1000) *1000
    plt.yticks(np.arange(0, max_price+graid, graid))
    plt.grid()
    plt.legend(loc = 'lower right')
    plt.savefig('./image/'+ str(title))
    plt.tight_layout() 
    plt.show()
    
    return
```

### 2.4.2 메인 함수 선언

```python
def main_function (df):
    
    scaled_stock, scaler_feature, scaler_label = min_max_scaler(df)

    train_size = 60 
    test_size = 20 
    
    train_data_set, test_data_set = divide_set(scaled_stock, train_size = train_size, test_size = test_size)
    
    n_steps_in = train_size 
    n_steps_out = test_size 
    step_size = 1
    
    X_train_full, y_train_full = split_sequences(train_data_set, n_steps_in = n_steps_in, n_steps_out = n_steps_out, step_size = step_size)
    X_test, y_test = split_sequences(test_data_set, n_steps_in = n_steps_in, n_steps_out = n_steps_out, step_size = step_size)
    
    n_features = X_train_full.shape[2]
    test_size = 0.2
    
    X_train, X_val, y_train, y_val = splitting_train_valid (X_train_full, y_train_full, test_size = test_size)
    
    model = set_lstm (X_train_full, n_steps_out = n_steps_out)

    history, y_pred, y_pred_origin, y_test_origin = model_fit (model, X_train, X_val, y_train, y_val, X_test, y_test, scaler_label, i, j)
    
    return i, j, history, y_pred, y_test, y_pred_origin, y_test_origin
```

```python
res_list_name = []
res_list = []

group_col_num = stock_array_set.shape[0]
stock_row_num = stock_array_set.shape[1]

for i in tqdm(range(group_col_num)) :
    for j in tqdm(range(stock_row_num)) :
        
        df_i_j = stock_array_set[i][j]

        i, j, history, y_pred, y_test, y_pred_origin, y_test_origin = main_function(df_i_j)
        
        set_lst = [i, j, history, y_pred, y_test, y_pred_origin, y_test_origin]
        
        res_list.append(set_lst)

print(len(res_list))
```

#### 2.5.3 예측 평가

##### 평가지표 함수 정의

```python
def MSE(y, pred):
    return np.mean(np.square(y-pred))

def RMSE(y, pred):
    return np.sqrt(np.mean(np.square(y-pred)))
```

```python
group_col_num = stock_array_set.shape[0]
stock_row_num = stock_array_set.shape[1]
element_num = np.array(res_list_copy).shape[-1]

result_array = np.array(res_list_copy).reshape(group_col_num, stock_row_num, element_num)

mse_scale_list = []
rmse_scale_list = []
mse_origin_list = []
rmse_origin_list = []

for i in range(group_col_num) :
    for j in range(stock_row_num) :
        
        y_pred = result_array[i][j][3]
#         print(y_pred.shape)
        y_test = result_array[i][j][4]
#         print(y_test.shape)
        y_pred_origin = result_array[i][j][5]
#         print(y_pred_origin.shape)
        y_test_origin = result_array[i][j][6]
#         print(y_test_origin.shape)
        
        df_y_pred_scaled = pd.DataFrame(y_pred)
        df_y_test_scaled = pd.DataFrame(y_test)
        df_y_pred_origin = pd.DataFrame(y_pred_origin)
        df_y_test_origin = pd.DataFrame(y_test_origin)        
        
        mse_scale = list(MSE(df_y_test_scaled, df_y_pred_scaled))
        rmse_scale = list(RMSE(df_y_test_scaled, df_y_pred_scaled))        
        mse_orig = list(MSE(df_y_test_origin, df_y_pred_origin))
        rmse_orig = list(RMSE(df_y_test_origin, df_y_pred_origin))
        

        mse_scale_list.append(mse_scale)
        rmse_scale_list.append(rmse_scale)        
        mse_origin_list.append(mse_orig)
        rmse_origin_list.append(rmse_orig)

```

```python
df_pred_mse = pd.DataFrame({'Pred_MSE' : np.squeeze(mse_scale_list)*100})
# df_pred_mse
```
