### 1.1 RAW DATA LOAD

#### 데이터 불러오기 및 확인

```python
df_raw = pd.read_csv('./dataset/v6_lowdata.csv', encoding='utf-8')
df = df_raw.copy()
```

#### 데이터 타입 확인 및 변경

```python
df['ticker'] = df['ticker'].map('{:06d}'.format)
df
```

### 1.2 데이터 전처리

#### 데이터 전처리 과정 확인

***********계년도 19년 이전 종목 삭제
* 상장일 2020-06-01 이후 상장한 종목 삭제
* 스팩주 삭제
* 시가총액 기준 200억 이상 종목 추출

#### 데이터 전처리 함수 생성

```python
df_pre = pd.DataFrame()
df_pre = df.copy()

def preprocessing_stock_selection() :
    ### 회계년도 19년 이전 종목 삭제
    df_pre = df[df['fiscal_year'] > 2019]
    
    ### 상장일 20200601 이후 상장한 종목 삭제
    df_pre['listingdate'] = df['listingdate'].str.replace('-','')
    df_pre['listingdate'] = df_pre['listingdate'].astype(int)
    df_pre = df_pre[df_pre['listingdate'] < 20190101]
    
    ### 스팩주 삭제
    df_pre = df_pre.drop(df_pre[df_pre['company_name'].str.contains('기업인수목적')].index, axis=0)
    
    ### 시가총액 기준 200억 이상 종목 추출
    df_pre['market_cap'] = df_pre['market_cap'] / 1000000.0  # 단위 100만 단위로 변경
    df_pre =  df_pre[df_pre['market_cap']  >= 20000.0]
    
    
    return df_pre
```

```python
df_psc = preprocessing_stock_selection().copy()
df_psc
```

### 1.3 재무/ 모멘텀/ 밸류 지표기준 종목순위 확인

#### 1.3.1 재무/ 모멘텀/ 밸류 지표기준별 Column_list 정리

```python
df_psc_col_list = df_psc.columns.to_list()
df_psc_col_list_base = df_psc_col_list[:5]
df_psc_col_list_fin = df_psc_col_list[5:11]
df_psc_col_list_mom = df_psc_col_list[11:20]
df_psc_col_list_value = df_psc_col_list[20:]
```

#### 1.3.2 재무/ 모멘텀/ 밸류 지표기준별 Dataframe 정리

```python
df_base_raw = df_psc[df_psc_col_list_base].copy()
df_finance_raw = df_psc[df_psc_col_list_fin].copy()
df_momentum_raw = df_psc[df_psc_col_list_mom].copy()
df_value_raw = df_psc[df_psc_col_list_value].copy()
```

#### 1.3.3 밸류 데이터 변환

```python
# 순서 정하는데 필요한 value_feature값 처치
# 'PER', 'PER_cal', 'PBR', 'PCR', 'PSR' 값 역수 변환

df_value_trans = df_value_raw.copy()
inverse_col_list = df_psc_col_list_value[:-1]


for inv_list in range(len(inverse_col_list)) :
    
    df_value_trans[inverse_col_list[inv_list]+"_inv"] = np.nan
    
    for i in range(len(df_value_raw)):
        
        if df_value_trans[inverse_col_list[inv_list]].iloc[i] == 0 :
            df_value_trans[inverse_col_list[inv_list]+"_inv"][i] = 0

        else :
            df_value_trans[inverse_col_list[inv_list]+"_inv"][i] = (1/(df_value_raw[inverse_col_list[inv_list]].iloc[i]))
            
```

```python
# 새로운 밸류 지표 리스트
df_psc_col_list_new_value = df_value_trans.columns.to_list()[5:]
df_psc_col_list_new_value
```

#### 1.3.4 시총기준 정규화 Feature 추가

```python
df_base = df_base_raw.copy()

# 1차 log Normalization
df_base['market_cap_log'] = df_base['market_cap'].apply(lambda x: np.log(x) if x != 0 else 0)

# 2차 Min-Max Normalization
    # min_val = df_base['market_cap_log'].min() 
    # max_val = df_base['market_cap_log'].max()
    # diff_val = max_val - min_val

lst = df_base['market_cap_log'].to_list()

def min_max_normalize(lst):
    normalized = []
    
    for value in lst:
        normalized_num = (value - min(lst)) / (max(lst) - min(lst))
        normalized.append(normalized_num)
    
    return normalized

```

```python
normalized = min_max_normalize(lst)
```

```python
base_col_len = len(df_base.columns.to_list())
df_base = df_base.assign(market_cap_MinMax = normalized)
df_base['market_cap_MinMax'] = df_base['market_cap_MinMax'].round(3).mul(100) # 0 ~ 100 사이 값으로 변경
```

#### 1.3.5 각 기준 종목순위 확인 함수 생성

```python
df_finance = df_finance_raw.copy()
df_momentum = df_momentum_raw.copy()
df_value = df_value_trans.copy()



df_rank_fin = pd.DataFrame()
df_rank_mom = pd.DataFrame()
df_rank_val = pd.DataFrame()

def set_rank_by_finance():

    for rank_list in range(len(df_psc_col_list_fin)) :

        if rank_list == df_psc_col_list_fin.index('debt_ratio'):
            df_rank_fin[df_psc_col_list_fin[rank_list]+"_rank"] = df_finance[df_psc_col_list_fin[rank_list]].rank(method='max', ascending=True)

        else :
            df_rank_fin[df_psc_col_list_fin[rank_list]+"_rank"] = df_finance[df_psc_col_list_fin[rank_list]].rank(method='max', ascending=False)
    
#     df_rank_fin['val_avg'] = np.nan
    
    df_rank_fin['fin_avg'] = df_rank_fin.iloc[:,:].mean(axis=1)

    return df_rank_fin   

def set_rank_by_momentum():
    
    for rank_list in range(len(df_psc_col_list_mom)) :
        
        if 'price' in df_psc_col_list_mom[rank_list] :
            pass

        else :
            df_rank_mom[df_psc_col_list_mom[rank_list]+"_rank"] = df_momentum[df_psc_col_list_mom[rank_list]].rank(method='max', ascending=False)

#     df_rank_mom['val_avg'] = np.nan
    
    df_rank_mom['mom_avg'] = df_rank_mom.iloc[:,:].mean(axis=1)

    return df_rank_mom

def set_rank_by_value():

    for rank_list in range(len(df_psc_col_list_new_value)) :
        
        df_rank_val[df_psc_col_list_new_value[rank_list]+"_rank"] = df_value[df_psc_col_list_new_value[rank_list]].rank(method='max', ascending=False)
        
    
    df_rank_val['val_avg'] = np.nan
    
    for i in range(len(df_value)):
        df_per_select = df_rank_val.iloc[i,:]
        df_disuse_cal = df_per_select.drop("PER_cal_inv_rank")
        df_rank_val['val_avg'][i] = df_disuse_cal.mean(axis=0)
    
    return df_rank_val    
```

##### 재무기준 종목순위

```python
df_rank_finance = set_rank_by_finance()
df_rank_finance = pd.merge(df_base, df_rank_finance ,left_index=True, right_index=True)
df_rank_finance
```

##### 모멘텀 기준 종목순위

```python
df_rank_momentum = set_rank_by_momentum()
df_rank_momentum = pd.merge(df_base, df_rank_momentum ,left_index=True, right_index=True)
df_rank_momentum
```

##### 밸류 기준 종목순위

```python
df_rank_value = set_rank_by_value()
df_rank_value = pd.merge(df_base, df_rank_value ,left_index=True, right_index=True)
df_rank_value
```

### 1.4 재무/모멘텀 기준 통합 순위 확인

```python
df_rank_fin_mom = pd.merge(df_base, df_rank_finance['fin_avg'], left_index=True, right_index=True)
df_rank_fin_mom = pd.merge(df_rank_fin_mom, df_rank_momentum['mom_avg'], left_index=True, right_index=True)
df_rank_fin_mom = pd.merge(df_rank_fin_mom, df_rank_value['val_avg'], left_index=True, right_index=True)
df_rank_fin_mom['fin_mom_sum'] = df_rank_fin_mom.iloc[:, -3:-1].sum(axis=1)
df_rank_fin_mom['sum_rank'] = df_rank_fin_mom['fin_mom_sum'].rank(method='max', ascending=True)
df_rank_fin_mom
```

#### 1.4.1 통합 순위 상위 50% 선별

```python
df_rank_sum_sort = df_rank_fin_mom.sort_values(by = 'sum_rank', ascending='True').copy()
df_rank_sum_sort
```

```python
df_rank_sum_half = df_rank_sum_sort.iloc[:int(len(df_rank_sum_sort)/2)].copy()
df_rank_sum_half = df_rank_sum_half.sort_values(by = 'market_cap', ascending = False)
df_rank_sum_half
```

### 1.6 대형주/중소형주 & 저평가/고평가 분류 
- A_group : 대형주 & 저평가(밸류 상위20위)
- B_group : 대형주 & 고평가(밸류 하위20위)
- C_group : 중소형주 & 저평가(밸류 상위20위)
- D_group : 중소형주 & 고평가(밸류 하위20위)

#### 1.6.1 A,B,C,D 그룹핑  

##### 각 그룹별 그룹핑 함수 생성

```python
df_A_group = pd.DataFrame()
df_B_group = pd.DataFrame()
df_C_group = pd.DataFrame()
df_D_group = pd.DataFrame()

def set_group(name):
    
    if name == 'A':
        df_var_sort = df_large_stocks.sort_values(by = 'val_avg', ascending = True).copy()
        df_A_group = df_var_sort[:20]
        return df_A_group
    
    elif name == 'B':
        df_var_sort = df_large_stocks.sort_values(by = 'val_avg', ascending = True).copy()
        df_B_group = df_var_sort[-20:]
        return df_B_group
    
    elif name == 'C':
        df_var_sort = df_small_stocks.sort_values(by = 'val_avg', ascending = True).copy()
        df_C_group = df_var_sort[:20]
        return df_C_group    

    elif name == 'D':
        df_var_sort = df_small_stocks.sort_values(by = 'val_avg', ascending = True).copy()
        df_D_group = df_var_sort[-20:]
        return df_D_group    
    
```

##### A_Group

```python
name = 'A'
df_A_group = set_group(name).copy()
df_A_group
```

##### B_Group

```python
name = 'B'
df_B_group = set_group(name).copy()
df_B_group
```

##### C_Group


#### 1.6.3 그룹별 종목 추출

```python
def ticker_by_group():

    group_name = ['A', 'B', 'C', 'D']
    df_ticker_set = pd.DataFrame() 
    tikcer_col = pd.DataFrame() 

    for name in group_name:
    #     print(set_group(name)['ticker'])

        tikcer_list = set_group(name)['ticker'].to_list()
        col_name = f'{name}_group'
        tikcer_col = pd.DataFrame({col_name: tikcer_list})
    #     tikcer_col.reset_index(drop=True, inplace=True)
    #     df_set_group.reset_index(drop=True, inplace=True)
        df_ticker_set = pd.concat([df_ticker_set, tikcer_col], axis = 1)
    
    return df_ticker_set
```

```python
df_ticker_set = ticker_by_group()
df_ticker_set
```
